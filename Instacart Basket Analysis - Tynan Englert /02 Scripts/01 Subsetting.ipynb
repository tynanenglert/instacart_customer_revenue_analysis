```python
# Step 3 of subsetting query is creating a new dataframe from that query 
df_snacks = df_prods[df_prods['department_id']==19]
```

```python
df_snacks.head()
```

```python
# Alternative to original subsetting query using loc function
df_snacks_2 = df_prods.loc[df_prods['department_id'] == 19]
```

```python
df_snacks_2
```

```python
# Other alternative to original subsetting query using loc and 'isin' - HELPFUL FOR LISTS
df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([19])]
```
